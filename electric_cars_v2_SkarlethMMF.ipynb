Pero te propongo generar un archivo como el anterior, pero que use el punto y coma como separador, en vez de la coma:

Para ello, debes diseñar una función que opera con un archivo como el de partida, que usa la coma como separador, y genera otro, con el punto y coma como separador.

```python
# Esta celda debe ser completada por el estudiante.

def to_semicolon(DatosComas, DatosPunComas):
    """
    DatosComas : Archivo de origen
    DatosPunComas: Nombre del archivo destino
    
    Esta funcion, lee in archivo de origen en formato csv, es decir cuyo separador es una coma , 
    y posteriormente escribe ese archivo usando como separador un punto y coma ; en el archivo destino.
    El archivo DatosComas debe estar almacenado en la misma carpeta que el Notebook.
    
    """
    
    docnew = ''
    f = open(DatosComas,'r')
    for linea in(f):
        linea = linea.replace(',',';')
        docnew = docnew+linea
    f.close()
    
    f2 = open(DatosPunComas,'w')
    f2.write(docnew)
    f2.close()
    
```

```python
# Ejecución de la función anterior:

DatosComas = "Electric_Vehicle_Population_Data.csv"
DatosPunComas = "Electric_Vehicle_Population_Data_pc.csv"
to_semicolon(DatosComas, DatosPunComas)
```

## a.2) Selección de una línea, separando sus campos

Diseña ahora una función que selecciona una línea y nos da una lista con los valores de sus campos. Los ejemplares de funcionamiento te darán la información sobre cómo deseamos que funcione:

```python
# Esta celda debe ser completada por el estudiante.
def select_line(DatosPunComas, linea):
    """
    DatosPunComas: es el nombre del archivo csv o txt, columnas separadas por ;
    linea: numero de linea que desea ver.
    Retorna la la linea pedida, con el caracter \n suprimido.
    """ 
    lineNumber = 0
    listLine = []
    
    with open(DatosPunComas, "r") as f:
        for line in f:
            if (lineNumber == linea):  
                listLine = (line.strip()).split(';')
                #print (listLine)
            lineNumber += 1 
        f.close()
    return listLine
```

## a.3) Coordenadas

La cadena `'POINT (-81.80023 24.5545)'` indica las coordenadas de un vehículo. Desearíamos convertirla en una función que da, para esa cadena, el par de reales siguiente: `(-81.80023 24.5545)`.

Pero cuidado: si la cadena fuera una lista vacía u otra cadena con un formato distinto, el valor generado debería ser `(-1, -1)`.

```python
# Esta celda debe ser completada por el estudiante.
def coords(cadena):
    """
    Recibe una cadena con formato: 'POINT (-xx.xxxxx yy.yyyyy)'
    Retorna un par ordenado  '(-xx.xxxxx, yy.yyyyy)'
    Excepcion: para 'nan' o cadenas vacias retorna '(-1, -1)'
    """
    punto = (-1,-1)
    if (str(cadena) =='nan') or len(cadena)<1: 
        punto = (-1,-1)
    else:
        if (cadena[0:5]=='POINT'):
            newCad = cadena[7:-1]
            cadSplited = newCad.split(' ')
            punto = (float(cadSplited[0]),float(cadSplited[1]))
            #print (punto)
    return punto
```

### b.3) Cálculo de la autonomía media de los vehículos en cada año

Diseña una función que calcule la autonomía media (redondeado a dos decimales)
de los vehículos registrados en cada año, a partir del archivo `csv` con punto y comas.

Las autonomías de vehículos consignadas con el valor cero no deben tenerse en cuenta, al tratarse de datos missing en realidad.

```python
# Esta celda debe ser completada por el estudiante.


# Funcion para leer la cabecera de la tabla, para encontrar la columna correspondiente a Autonomia. 

def findcolumn(cadena):
    """
    Esta funcion encuentra el indice de la columna correspondiente 
    a la cadena dada como parametro de entrada.
    """
    cabecera = select_line(DatosPunComas, 0)
    #print(cabecera)
    lc = len(cabecera)
    for i in range(lc):
        if cabecera[i]== cadena:
            return i
    return -1


def aut_medias(DatosPunComas):
    """
    DatosPunComas: archivo csv con punto y comas.
    Esta funcion retorna la autonomía media (redondeado a dos decimales) de los vehículos registrados en cada año, 
    a partir del archivo DatosPunComas.
    """
    index1 = findcolumn('Model Year')
    index2 = findcolumn('Electric Range')
    Autonomias = dict()
    Dic_Years = dict()
    
    with open(DatosPunComas, "r") as f:
        linea = f.readline()
        #print(linea)
        i=0
        for line in f:
            i +=1
            elementos = (line.strip()).split(';') 
            Autonomias [elementos[0]+str(i)] = { 'Year': elementos[index1], 
                                         'Electric Range': float(elementos[index2])}
    f.close()
    #print (len(Autonomias))
    #print (' ',i)
    
    for i, er in Autonomias.items():
        #print (er)
        y = int(er['Year'])
        erval = er['Electric Range'] 
        if (erval>0):
            Dic_Years[y] = [erval] if y not in Dic_Years.keys() else Dic_Years[y] + [erval]
    
    for y, er in Dic_Years.items():
        mean_er = np.mean(er)
        Dic_Years [y] = np.round((mean_er),2) 

    #print(Dic_Years)
    return Dic_Years

#autonom_medias =  aut_medias(DatosPunComas)
#print(autonom_medias)
```

### c.1 Un modelo de gráfica

Vamos a diseñar un modelo de gráfica sencillo que nos sirva para las siguientes representaciones. Tomará como parámetro una lista de pares $(x, y)$, y opcionalmente los tres rótulos explicativos que necesitamos incluir. Además, queremos que las etiquetas de las abcisas aparezcan inclinadas, para poder luego mostrar intervalos de edad.

Las pruebas de funcionamiento te darán más información que las explicaciones que pueda yo dar aquí.  

```python
# Esta celda debe ser completada por el estudiante.

def representar_xxx_yyy(arrayP, titulo = ['','','']): # titulos opcionales
    """
    Recibe:
        arrayP: Arreglo de pares ordenados a graficar.
        titulo: Cadena titulo del grafico. [Opcional]
    
    Funcion crea un grafico a partir de arrayP.
    Muestra grafico.
        
    """
    a = np.array(arrayP)
    x = np.transpose(a)[0]
    y = np.transpose(a)[1] 
    plt.figure(figsize=(12,4))
    plt.plot( y,'-')
    plt.title(titulo[0])
    plt.ylabel(titulo[1])
    plt.xlabel(titulo[2]) 
    plt.grid(True)
    plt.xscale('linear')
    x_pos = np.arange(len(x))
    x_labels = [int(i) for i in x]
    plt.xticks(x_pos, x_labels, rotation=45)
    plt.show()
    
```

```python
# Pruebas de funcionamiento:

representar_xxx_yyy([(1, 8), (2, 4), (3, 2), (4, 1), (5, 0.5), (6, 0.25)], ["Serie descendente", "Ordenadas", "Abcisas"])

representar_xxx_yyy([(1, 1), (2, 2), (3, 4), (4, 8), (5, 16), (6, 32)])
```

```python
def sepLatLong(column_coors):
    """
    Esta funcion recibe la columna con las coordenadas geograficas en 
    formato (-+x.xxx, -+y.yyy), y las separa en dos columnas separdas, 
    para latitud y longitud. 
    Retorna dos arreglos, en el orden de latitud y longitud.
    
    """
    ccl = len(column_coors)
    lat = np.zeros(ccl)
    lon = np.zeros(ccl)
    i=0
    for v in (column_coors): 
        lat[i], lon[i] = float(v[0]), float(v[1])
        #print (lat[i], lon[i])
        i +=1
    return lat, lon

ccl = len(tabla)
tabla['Latitude'] =  np.zeros(ccl)
tabla['Longitude'] =  np.zeros(ccl)

tabla['Latitude'], tabla['Longitude'] = sepLatLong(tabla['Vehicle Location'])
```

```python

gdf = geopandas.GeoDataFrame(
    tabla, geometry = geopandas.points_from_xy(
    tabla.Latitude, tabla.Longitude))

world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

```python
# Pruebas de funcionamiento:
fig, ax = plt.subplots(figsize=(20, 20))
world[world.continent == 'North America'].plot(ax = ax,
    figsize=(15,15), alpha = 0.4,
    color='gray', edgecolor='black')

gdf.plot(column = "Make", cmap='Blues', edgecolor='blue', ax=ax,  
    legend = True, legend_kwds={'loc':'lower right', 'fontsize':'14'})
plt.title('Ubicacion de las diferentes marcas', fontsize= 32)  
plt.show()
```
